```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Churn_Modelling.csv')
df
```

```python
#Creating indepemdent or dependent features-

x=df.iloc[:,3:13]
y=df.iloc[:,13]
```

```python
x.head()
```

```python
y
```

```python
#feature encoding-

Geography=pd.get_dummies(x['Geography'],drop_first=True)
Gender=pd.get_dummies(x['Gender'],drop_first=True)
```

```python
x=x.drop(columns=['Geography','Gender'],axis=1)
x
```

```python
x=pd.concat([x,Geography,Gender],axis=1)
x
```

```python
#Train & test split-

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Feature scaling-

from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train.shape
```

```python
#Lets create ANN model-

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU,ReLU
from tensorflow.keras.layers import Dropout
```

```python
#Lets initialize ANN-

classifier=Sequential()
```

```python
#Adding input layer-

classifier.add(Dense(units=11,activation='relu'))
```

```python
# Adding first hidden layer- also using drop out layer as regularization

classifier.add(Dense(units=7,activation='relu'))
classifier.add(Dropout(0.2))
```

```python
#Adding second hidden layer-

classifier.add(Dense(units=6,activation='relu'))
classifier.add(Dropout(0.3))
```

```python
#Adding output layer-

classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
#Model training-

import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
classifier.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Early stopping- Go to early stopping keras to get the code for early stopping

# Early stopping is a process of stopping the training of ANN model when accuracy/loss becomes stagnent or does not increases.
```

```python
early_stopping=tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
model_history=classifier.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=1000,callbacks=early_stopping)
```

```python
model_history.history.keys()
```

```python
#Summarize history for accuracy-
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
#Here we can see that after EPOCH 32 the model training stopped as there was no further change in accuracy/loss.
```

```python
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
#Prediction for test data-
```

```python
y_pred=classifier.predict(x_test)
y_pred=(y_pred>=0.5)
```

```python
#Confusion matrics-

from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```
